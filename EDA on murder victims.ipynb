```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
# Missing/Null values
# In this dataset nan => 0

df = df.replace(np.nan,0)
df.head()
```

```python
df= df[['Area', 'Year', 'Group_Name', 'Sub_Group_Name','Victims_Upto_10_Yrs','Victims_10_to_15_Yrs','Victims_15_to_18_Yrs', 'Victims_18_to_30_Yrs','Victims_30_to_50_Yrs','Victims_Above_50_Yrs', 'Victims_Total']]
df.head()       
```

```python
# Correlation

df.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.heatmap(df.corr(),annot=True,cmap="Blues")
plt.title("Correlation")
plt.show()
```

```python
# Area wise victims

plt.figure(figsize=(14,6))
a=sns.barplot(x=df["Area"],y=df["Victims_Total"])
a.set_xticklabels(a.get_xticklabels(), rotation=90, ha="right")
plt.title("Victims in each area")
plt.tight_layout()
plt.show()
```

```python
# Victims over the years

sns.barplot(x=df["Year"],y=df["Victims_Total"])
plt.title("Victims over the years")
plt.show()
```

```python
# Avoiding repetation in Group_Name by removing Murder-Total Victims category

df = df[df.Group_Name != 'Murder - Total Victims']
df.head()
```

```python
# Victims distribution over gender

sns.barplot(x=df["Group_Name"],y=df["Victims_Total"])
plt.title("Gender wise distribution")
plt.show()
```

```python
df1 = df.groupby(['Year'])['Victims_Upto_10_Yrs','Victims_10_to_15_Yrs','Victims_15_to_18_Yrs', 'Victims_18_to_30_Yrs','Victims_30_to_50_Yrs','Victims_Above_50_Yrs'].sum().reset_index()
df1.head()
```

```python
df1 = df1.melt(id_vars =['Year'],value_vars=['Victims_Upto_10_Yrs','Victims_10_to_15_Yrs','Victims_15_to_18_Yrs', 'Victims_18_to_30_Yrs','Victims_30_to_50_Yrs','Victims_Above_50_Yrs'],var_name = "Age_category")
df1.head()
```

```python
# Gender wise distribution of total victims

a=sns.barplot(x=df1["Age_category"],y=df["Victims_Total"])
a.set_xticklabels(a.get_xticklabels(), rotation=90, ha="right")
plt.title("Gender wise distribution of Total victims")
plt.tight_layout()
plt.show()
```
